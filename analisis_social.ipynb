### 1.1 — Variables: información sobre el autor

Cada notebook tiene que identificar quién lo hizo.
Usamos **variables** para guardar esa información.
Una variable es un nombre que le damos a un valor para poder reutilizarlo.

```python
# Datos del autor — estos son los del ejemplo
# Los tipos de datos que usamos:
#   str  → texto entre comillas
#   int  → número entero

nombre_autor = "vanesameinardi"
legajo       = 1374
provincia_de_interes = "Córdoba"

print(f"Notebook preparado por: {nombre_autor}")
print(f"Legajo: {legajo}")
print(f"Provincia de interés: {provincia_de_interes}")
```

### 1.2 — Listas: datos de varias provincias

Una **lista** agrupa varios valores bajo un mismo nombre.
Cada posición de la lista corresponde a una provincia — el orden importa.
El índice 0 es Buenos Aires, el 1 es Córdoba, y así.

```python
# Listas con datos sociales ficticios
# Fuente: valores aproximados basados en informes del INDEC 2022

provincias     = ["Buenos Aires", "Córdoba", "Santa Fe", "Tucumán", "Chaco"]
tasa_pobreza   = [38.2, 34.5, 36.1, 48.3, 57.8]   # porcentaje
tasa_desempleo = [11.5, 10.2, 10.8, 14.1, 16.5]   # porcentaje
acceso_agua    = [95.2, 93.1, 91.8, 82.4, 68.3]   # porcentaje con acceso

print("Datos cargados:")
print(f"  Provincias    : {provincias}")
print(f"  Tasa pobreza  : {tasa_pobreza}")
print(f"  Tasa desempleo: {tasa_desempleo}")
print(f"  Acceso a agua : {acceso_agua}")
```

### 1.3 — Loops: recorrer los datos

Un **loop** nos permite repetir una acción para cada elemento de una lista.
`range(len(provincias))` genera los índices 0, 1, 2, 3, 4 —
uno por cada provincia — y `i` toma ese valor en cada vuelta del loop.

```python
# Recorremos todas las provincias y mostramos sus datos
print(f"{'Provincia':<15} {'Pobreza':>10} {'Desempleo':>11} {'Agua':>8}")
print("-" * 48)

for i in range(len(provincias)):
    # i toma los valores 0, 1, 2, 3, 4 en cada vuelta
    # provincias[i] accede al elemento i de la lista
    print(f"{provincias[i]:<15} {tasa_pobreza[i]:>9.1f}% {tasa_desempleo[i]:>10.1f}% {acceso_agua[i]:>7.1f}%")
```

### 1.4 — Condicionales: clasificar los datos

Un **condicional** ejecuta código distinto según si una condición es verdadera o falsa.
Acá lo usamos para clasificar cada provincia según su nivel de pobreza.

```python
# Clasificamos cada provincia según su tasa de pobreza
# Si pobreza < 40%  → situación moderada
# Si pobreza < 50%  → situación alta
# Si pobreza >= 50% → situación crítica

print("Clasificación por tasa de pobreza:")
print()

for i in range(len(provincias)):
    pobreza = tasa_pobreza[i]

    if pobreza < 40:
        nivel = "🟡 Moderada"
    elif pobreza < 50:
        nivel = "🟠 Alta"
    else:
        nivel = "🔴 Crítica"

    print(f"  {provincias[i]:<15}: {pobreza:.1f}%  →  {nivel}")
```

### 1.5 — Funciones: reutilizar el código

Una **función** es un bloque de código con nombre que podemos llamar cada vez que lo necesitemos.
Acá definimos una función que calcula el promedio de cualquier lista de números.

```python
# Definimos la función calcular_promedio
# Recibe una lista de números y devuelve su promedio

def calcular_promedio(lista):
    total = 0
    for valor in lista:
        total = total + valor
    promedio = total / len(lista)
    return promedio

# Llamamos a la función con cada indicador
prom_pobreza   = calcular_promedio(tasa_pobreza)
prom_desempleo = calcular_promedio(tasa_desempleo)
prom_agua      = calcular_promedio(acceso_agua)

print("Promedios nacionales (muestra de 5 provincias):")
print(f"  Pobreza    : {prom_pobreza:.1f}%")
print(f"  Desempleo  : {prom_desempleo:.1f}%")
print(f"  Acceso agua: {prom_agua:.1f}%")
```

## Parte 4 — Explorar el trabajo de todos (lo hacemos juntos)

Una vez que todos hayan subido sus cambios, vamos a explorar los forks de la clase juntos.

```python
# Ver los forks del repositorio desde Python
# (requiere conexión a internet — ejecutar en clase)

import urllib.request
import json

# Reemplazá con el usuario real del docente
usuario_docente = "usuario-docente"
repo            = "intro-programacion-datos"

url = f"https://api.github.com/repos/{usuario_docente}/{repo}/forks"

try:
    with urllib.request.urlopen(url) as response:
        forks = json.loads(response.read())

    print(f"Forks del repositorio '{repo}':")
    print(f"Total: {len(forks)}")
    print()
    for fork in forks:
        print(f"  → {fork['owner']['login']}")
        print(f"     {fork['html_url']}")
except Exception as e:
    print(f"No se pudo conectar: {e}")
    print("Asegurate de reemplazar 'usuario-docente' con el nombre de usuario real.")
```
